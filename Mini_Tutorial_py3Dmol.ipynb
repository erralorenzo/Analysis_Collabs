## Visualizacion de una determinada cadena

````python
import py3Dmol
#Codigo de PDB que queremos visualizar
pdb='1mbo'
# Iniciamos el visualizador

view=py3Dmol.view(query=f'pdb:{pdb}')
#Zoom a la estructura
view.zoomTo()
#Color de fondo
view.setBackgroundColor('grey')
#Visualizamos la cadena A con el stylo de Cartoon en un color determinado
view.setStyle({'chain':'A'},{'cartoon': {'color':'purple'}})
#Inicializamos el visualizador
view.show()
````

## Visualizacion especifica

- de un determinado residuo
- de todos los aminoacidos de un tipo
- un atomo

````python
view=py3Dmol.view(query=f'pdb:{pdb}')
view.zoomTo()
view.setBackgroundColor('white')

# Para ver una posicion especifica vamos con el comando RESI
view.addStyle({'chain':'A','resi':58},{'stick':{'colorscheme':'purple'}})

# Para ver todos los aminoacidos Alanina  vamos con el comando RESN
view.addStyle({'chain':'A','resn':'ALA'},{'sphere':{'colorscheme':'greenCarbon'}})

# Para ver un Atomo especifico
view.setStyle({'elem': 'Fe'}, {'sphere': {'radius': 1, 'color': 'yellow'}})
````

## Visualizacion estructura con labels

````python
view.clear()

# Combinamos un poco lo visto para visualizar varias cadenas y le agregamos Labels
viewer = py3Dmol.view(query='pdb:5WOG')

viewer.addLabel('alpha subunits', {'fontColor':'orange', 'backgroundColor':'lightgray'},
                {'chain': ['A','B']})
viewer.setStyle({'chain':['A','B']},{'cartoon': {'color': 'red'}}) # alpha subunits of hemoglobin


viewer.addLabel('beta subunits', {'fontColor':'blue', 'backgroundColor':'lightgray'},
                {'chain': ['C','D']})
viewer.setStyle({'chain':['C','D']},{'cartoon': {'color': 'blue'}}) # beta subunits of hemoglobin

viewer.setStyle({'resn': 'HEM'},{'sphere': {'colorscheme': 'greenCarbon'}})
viewer.show()
````

## Visualizacion grupo hemo y residuos a una determinada distancia

````python
# Visualizamos los residuos a una determinada distancia del grupo hemo
distancia = 5
view = py3Dmol.view(query='pdb:1mbo')
view.clear()
# select by distance
selection = {'resn':'HEM', 'chain':'A', 'expand': distancia}

# set styles
view.setStyle(selection,{'stick':{'colorscheme':'orangeCarbon'}})
viewer.setStyle({'chain': 'A', 'resn': 'HEM'},{'sphere': {'colorscheme': 'lightgreenCarbon'}})
viewer.zoomTo(selection)
viewer.show()
````

## Visualizacion con con lineas


````python
import py3Dmol

# Cargar la estructura PDB
view = py3Dmol.view(query='pdb:1mbo')  # Cambia '1A2C' por el código PDB o tu archivo

# Estilo general de la estructura (por ejemplo, un estilo cartoon)
view.setStyle({'chain':['A']},{'cartoon': {'color':'purple'}})

# Seleccionar dos átomos específicos (por ejemplo, atomos de residuo 50 y 100)
# Ajusta 'resi' y 'atom' según tu estructura
atom1 = {'chain':'A','resi': 58, 'atom': 'CA'}
atom2 = {'chain':'A','resi': 100, 'atom': 'CA'}

# Añadir una línea entre los átomos para visualizar la distancia
view.addLine({'color':'red','linewidth': 100,'start':atom1,'end':atom2})

view.show()
````

# Acceder a tipos de elementos

Cambia la manera de referirse a los residuos y átomos en la estructura PDB, específicamente en si se trata de un residuo de la **proteína** o de un **heteroátomo** (molécula no estándar como un cofactor).

### 1. Seleccionar un átomo dentro de un residuo proteico:
   ```python
   atom = structure[0]["A"][100]["CA"] ```

  Aquí, 100 representa el número del residuo en la cadena 'A'.
  "CA" es el nombre del átomo dentro de este residuo, en este caso, el carbono alfa de un aminoácido (en una proteína).
  
  
  *Uso: Esta estructura es adecuada para seleccionar átomos en residuos proteicos estándar que forman parte de la cadena de la proteína.*


### 2. Seleccionar un átomo dentro de un heteroátomo:

   ```python
      atom_NA = structure[0]['A'][('H_HEM', 155, ' ')]['NA'] ```


  ('H_HEM', 155, ' ') se usa aquí para identificar un heteroátomo, en este caso, el residuo de hemo (H_HEM) en la posición 155 de la cadena 'A'.
  
  NA es el nombre específico del átomo dentro de este heteroátomo.
  
  *Uso: Esta estructura se utiliza para heteroátomos o moléculas no estándar en la estructura, como cofactores, ligandos, o iones, que no son parte de la secuencia proteica principal.*

En resumen, para seleccionar un átomo dentro de la proteína usa el número del residuo directamente, mientras que para heteroátomos se usa una tupla que identifica tanto el nombre del heteroátomo como su número de residuo y cualquier código de inserción.
